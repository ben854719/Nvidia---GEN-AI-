```python
import pandas as pd
```

```python
# Import the dataset from Nvidia
Stock_Nvidia = pd.read_csv('Stock_Nvidia.csv')
print(Stock_Nvidia.head(25))
```

```python
# The rows 10 to 13 are duplications in the dataset. I will remove them
Stock_Nvidia = Stock_Nvidia.drop_duplicates()
print(Stock_Nvidia.head(25))
```
